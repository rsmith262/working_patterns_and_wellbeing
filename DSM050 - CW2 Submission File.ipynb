## Introduction

The health benefits of a physically active lifestyle are well known. Studies show irrefutable evidence proving the prevention of many chronic diseases and premature death [1]. However, a study in 2018 says that a quarter of the world’s population is physically inactive [2].

This study aims to investigate the relationship between employment and exercise habits. I would like to see if factors in peoples working life affect the amount of exercise they undertake and whether there is a link between home working and physical activity. Furthermore, the study will look at whether taking time to exercise can improve wellbeing and productivity during the working day.

### Background Research

In the US more than 80% of jobs are now sedentary [3]. The rise of the internet has helped fuel sedentary lifestyles and working from home has become more prevalent removing the need for physical activity even further [4]. The move to homeworking was further fuelled by the Covid-19 pandemic in 2020 which now makes homeworking for at least part of the week common practice. ONS data shows 44% of workers reported to be hybrid or home workers in the period between September 2022 and January 2023. In early 2020 this figure was 49% showing that this shift has largely stuck since the pandemic [5].

Although it can be argued that homeworking gives workers a better work life balance [6], there is mixed evidence regarding whether it is beneficial to their physical activity, with one study showing 4 in 10 people surveyed had decreased physical activity as a result and 3 in 10 reporting that it had increased [7]. Other studies suggest homeworking can lead to more sedentary behaviour causing people to spend two more hours a day sitting [8]. Although commuting stress is decreased, it does mean people are moving less.

Research shows higher earners and those with the highest educational attainment are most likely to report working from home or flexible working [5]. Studies also suggest there is a positive correlation between educational attainment and increased physical activity [9,10].

Exercise during the working day can be linked to increased mood and workplace performance with workers reporting better concentration and less stress on days they exercised vs days they did not [11]. Socio-demographic factors have also been shown to have a significant effect on the amount of physical activity achieved [12].


### Motivation and Research Objectives

Consistent physical activity has many positive effects and plays a vital role in keeping us healthy. With nearly 20% of our total waking hours spent working [13] and evidence to suggest that lifestyles are becoming more sedentary, it would be beneficial to understand the affects our work and personal lives have on our physical activity.

The aim of this study is to find what attributes affect the level of physical activity in the working age population. It will assess the role work and education has on the level of physical activity attained, how homeworking is affecting the amount of physical activity employees are getting and how exercise effects energy, concentration, productivity, and stress.

It could be used for initiatives to target specific demographics in different ways to make sure approaches are tailored to subsets of the population. As well as to inform homeworking policies and guidance.

### Domain Concepts and External Data Source

Throughout the study I will be making use of some specific terminology:

Physical Activity: defined as physical movement generated by skeletal muscle usage that results in energy expenditure above the rest value [17].

Exercise: is a form of physical activity that is planned, structured, repetitive and is usually performed to gain physical benefits [18].

Sedentary behaviour: “Sedentary behaviour is any waking behaviour characterised by an energy expenditure ≤1.5 metabolic equivalents (METs), while in a sitting, reclining or lying posture” [19]. For example, computer use, watching television or driving.

Energy: The ability to do work, energy levels could refer to physical or mental energy.

Concentration: The ability to focus one’s attention and thought processes on a specific problem or subject [20].

Productivity: The output you can achieve over a specific timeframe. It can be measured in terms of units of work completed e.g., how many items delivered in an hour [21].

Stress: Stress can affect almost every system in the body and can influence feelings and behaviours. Some levels are normal reactions to everyday stimulation and pressures, but it becomes unhealthy when it has negative consequences on your day-to-day functioning [22].

Active commuting: Commuting by means of physical activity e.g., walking, running, or cycling [23].

```python
# Function to get column position for inserting new column
# I want to keep a similar structure to the data rather than new columns going to the end
def col_position(df,adj_col):
    """
    Function to get column position for inserting new column
    
    :df1: the dataframe being analysed
    :type: pandas dataframe
    
    :adj_col: the column to the left of where you want to place the new column
    :type: str
    
    :return: returns the location of the new column
    :type: int
    
    """
    return df.columns.get_loc(adj_col) + 1 # gets column position + 1
```

#### Missing and null values

Missing and null values could cause issues later so I'm going set out a strategy for dealing with them here.

During the time when our surveys were live the JISC online survey platform had some documented issues whereby not all data was being captured correctly, this caused a few missing pieces of data.

As I don't have a huge amount of survey responses I would rather not drop rows unless really necessary
I dealt with two columns earlier days_exercised and minutes_exercised by filling with the median value for the respective columns. For the other columns I will:

* drop columns with more than 80% missing data as these won't add any value
* fill work_hours_exercise and wfh_exercise_likelhood with mode value
* fill minutes_exercise with median
* I want to be able to compare te typical day (td) likert scales with the work hours exercise likerts (whe). Due to a glitch in the JISC system 1 respondent managed to enter results in whe_concentration and whe_stress but not the other two so I'm going to fill these with the values from the respondent's typical day so these can be compared later on. The other null values in this column will be left and filtered out when comparing the two for changes later.

```python
# checking for null values

cols_null_values = survey_data.columns[survey_data.isnull().any()].tolist() # list of cols with null values

# this prints the % of data missing from each column that has missing values
print((survey_data[cols_null_values].isna().mean()*100).sort_values(ascending=False))
```

```python
# dropping columns with more than 80% missing data

survey_data = survey_data.drop(columns=['other_company', 
                                        'exercise_type_other',
                                        'part_time_days',
                                        'ac_energy',
                                        'ac_concentration',
                                        'ac_productivity',
                                        'ac_stress'
                                        ])
```

```python
# fill work_hours_exercise and wfh_exercise_likelhood with mode value

# get the mode values
mode_work_hours_exercise = survey_data['work_hours_exercise'].mode().iloc[0]
mode_wfh_exercise_likelihood = survey_data['wfh_exercise_likelihood'].mode().iloc[0]
median_minutes_exercise = survey_data['minutes_exercise'].median()

# fills the missing values
survey_data['work_hours_exercise'].fillna(mode_work_hours_exercise, inplace=True)
survey_data['wfh_exercise_likelihood'].fillna(mode_wfh_exercise_likelihood, inplace=True)
survey_data['minutes_exercise'].fillna(median_minutes_exercise, inplace=True)
```

```python
# fill null whe_concentration and whe_stress with values from td_concentration and td_stress [27]
# https://www.geeksforgeeks.org/boolean-indexing-in-pandas/

# first whe_concentration
# finds row where whe_concetration is null and whe_energy is not null
null_condition = survey_data['whe_concentration'].isna() & survey_data['whe_energy'].notna()

# fills null in whe_concentration with value from td_concentration, only where the condition is met
survey_data.loc[null_condition, 'whe_concentration'] = survey_data.loc[null_condition, 'td_concentration']

# second whe_stress
# finds row where whe_concetration is null and whe_energy is not null
null_condition = survey_data['whe_stress'].isna() & survey_data['whe_energy'].notna()

# fills null in whe_concentration with value from td_concentration, only where the condition is met
survey_data.loc[null_condition, 'whe_stress'] = survey_data.loc[null_condition, 'td_stress']
```

```python
# dropping year_of_birth and completion_data as these are not needed

survey_data = survey_data.drop(columns=['year_of_birth', 
                                        'completion_date',])
```

```python
# data to keep:

# SEQN - Respondent sequence number
# PAQ610 - Number of days vigorous work - use these two to get a kind of 
# sedentary, lightly, highly active category 
# PAQ625 - Number of days moderate work - change outliers to nan then do
# if vig greater than mod then high, if mod higher than vig then lightly, else sedentary
# PAQ650 - Vigorous recreational activities
# PAQ655 - Days vigorous recreational activities
# PAD660 - Minutes vigorous recreational activities
# PAQ665 - Moderate recreational activities
# PAQ670 - Days moderate recreational activities
# PAD675 - Minutes moderate recreational activities

# takes only relevant columns
df_pa = df_pa[['SEQN','PAQ610','PAQ625','PAQ650','PAQ655','PAD660','PAQ665','PAQ670','PAD675']].copy()

# rename column headers to sensible names
df_pa.rename({
    'PAQ610': 'days_vigorous_work',
    'PAQ625': 'days_moderate_work',
    'PAQ650': 'vigorous_rec_y/n',
    'PAQ655': 'days_vigorous_rec',
    'PAD660': 'minutes_vigorous_rec',
    'PAQ665': 'moderate_rec_y/n',
    'PAQ670': 'days_moderate_rec',
    'PAD675': 'minutes_moderate_rec' 
    }, axis=1,inplace=True)

df_pa.head()
```

```python
# joining the dfs

# number of records
print(f'df_demo has {df_demo.shape[0]} records')
print(f'df_occ has {df_occ.shape[0]} records')
print(f'df_pa has {df_pa.shape[0]} records')
matching_records = min(df_demo.shape[0],df_occ.shape[0],df_pa.shape[0])
print(f'\nAfter merging there should be {matching_records} '
      f'records as we want to keep only those with matching records in each set:')
# merging the df's together using 'SEQN' as the key
external_data = pd.merge(df_demo,df_occ,on='SEQN').merge(df_pa,on='SEQN')

# number of records in merged set
print(f'\ndf_merged has {external_data.shape[0]} records')
```

```python
# map column values to something more meaningful

# gender
gender_mapping_dict = {
    1: 'Male',
    2: 'Female'
}

external_data['gender'] = external_data['gender'].map(gender_mapping_dict)


# employment status - full/part time
# classing people that worked 35 hours+ as full time

emp_status_mapping_dict = {
    1: 'Full Time',
    2: 'Part Time',
    7: np.nan,
    9: np.nan
}
external_data['work_35+_hours'] = external_data['work_35+_hours'].map(emp_status_mapping_dict)

# education
education_mapping_dict = {
    1: 'Less than 9th grade',
    2: '9-11th grade',
    3: 'High school graduate',
    4: 'Some college or AA degree',
    5: 'College graduate or above',
    7: np.nan,
    9: np.nan
}
external_data['education'] = external_data['education'].map(education_mapping_dict)

# marital_status
marital_mapping_dict = {
    1: 'Married/Living with Partner',
    2: 'Widowed/Divorced/Separated',
    3: 'Never married',
    77: np.nan,
    99: np.nan
}
external_data['relationship_status'] = external_data['relationship_status'].map(marital_mapping_dict)
```

```python
# create age category column - making this the same as my survey

# age_cat_position = survey_data.columns.get_loc('age') + 1
external_data.insert(col_position(external_data,'age'), 'age_category',
                   
                   pd.cut(external_data['age'],bins=[18,25,35,45,55,65,100],\
                   labels=['18-24','25-34','35-44','45-54','55-64','65+'])                 
                  )
```

```python
#prefered exercise intensity
# if minutes vigorous > minutes moderate then High else low

# first map values to clean out missing values
ex_intensity_mapping_dict = {
    7777: np.nan,
    9999: np.nan
}

cols_to_map = ['minutes_vigorous_rec','minutes_moderate_rec']

# replace values in mapping dict
external_data[cols_to_map] = external_data[cols_to_map].replace(ex_intensity_mapping_dict)

# this defines the logic for the new column
def intensity_logic(row):
    if pd.isna(row['minutes_vigorous_rec']) and pd.isna(row['minutes_moderate_rec']):
        return np.nan
    elif pd.notna(row['minutes_vigorous_rec']) and pd.notna(row['minutes_moderate_rec']):
        return 'Mixed'
    elif pd.notna(row['minutes_vigorous_rec']):
        return 'High'
    else:
        return 'Moderate'

# apply the logic using apply() to create the new column
external_data['exercise_intensity'] = external_data.apply(intensity_logic, axis=1)
```

```python
# total minutes exercised

# defines logic
#logic is days (days_vigorous_rec * minutes_vigorous_rec) + (days_moderate_rec * minutes_moderate_rec)
def total_minutes_logic(row):
    if pd.isna(row['minutes_vigorous_rec']) and pd.isna(row['minutes_moderate_rec']):
        return np.nan
    elif pd.isna(row['minutes_vigorous_rec']):
        return row['days_moderate_rec'] * row['minutes_moderate_rec']
    elif pd.isna(row['minutes_moderate_rec']):
        return row['days_vigorous_rec'] * row['minutes_vigorous_rec']
    else:
        return (row['days_vigorous_rec'] * row['minutes_vigorous_rec']) + \
               (row['days_moderate_rec'] * row['minutes_moderate_rec'])

external_data['minutes_exercise'] = external_data.apply(total_minutes_logic, axis=1)
```

```python
def gender_pie(df, ax, data_set_str):
    
    # some ideas from here [31]
    # https://medium.com/@kvnamipara/a-better-visualisation-of-pie-charts-by-matplotlib-935b7667d77f
    
    # gets occurrences of each gender
    gender_count = df['gender'].value_counts().reindex(['Male', 'Female']).dropna()
    
    # gets colours so they are consistent each time a plot is called
    colors = sns.color_palette('viridis', n_colors=2)

    # adjusts order so always 'Male' first, then 'Female'
    ordered_counts = gender_count[['Male', 'Female']]
    
    # custom labels for adding a count
    labels = [f"{label}: {count}" for label, count in ordered_counts.items()]

    # creates pie
    _, _, autotexts = ax.pie(ordered_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    
    # changes the label colours
    for autotext in autotexts:
        autotext.set_color('white')

    ax.set_title(f'Gender Distribution ({data_set_str})')
    # removes y axes label
    ax.set_ylabel('')
```

```python
def age_cat_count(df,data_set_str):
   # creates chart
    plt.figure(figsize=(15,6)) # setting the figure size
    # creates axes object
    ax = sns.countplot(x=df['age_category'],palette='viridis')

    for i in ax.containers: # this adds data labels
        ax.bar_label(i)

    # setting the labels and titles
    ax.set_xlabel('Age Category')
    ax.set_ylabel('Count')
    ax.set_title(f'Chart showing count of respondents across age categories ({data_set_str})')
    plt.show()
```

## 2.	Exercise during work hours has a positive effect on energy, concentration, productivity, and stress levels.

#### Code and Functions for Charts

```python
# create new dfs with just likert data
likert_df = survey_data.dropna().copy()
td_df = likert_df[['td_energy','td_concentration','td_productivity','td_stress']]
whe_df = likert_df[['whe_energy','whe_concentration','whe_productivity','whe_stress']]
```

```python
def likert_tables(td_df,whe_df):
    
    # mapping dictionary for renaming cols
    td_mapping_dict = {
        'td_energy': 'Energy',
        'td_concentration': 'Concentration',
        'td_productivity': 'Productivity',
        'td_stress': 'Stress'
    }
    
    whe_mapping_dict = {
        'whe_energy': 'Energy',
        'whe_concentration': 'Concentration',
        'whe_productivity': 'Productivity',
        'whe_stress': 'Stress'
    }
    
    #creates new local df and rename cols based on mapping dict
    td_df_local = td_df.rename(columns=td_mapping_dict)
    whe_df_local = whe_df.rename(columns=whe_mapping_dict)
    
    # melts data into longform [32]
    # https://pandas.pydata.org/docs/reference/api/pandas.melt.html
    td_df_melt = td_df_local.melt(var_name='Typical Day Likert', value_name='Rating')
    whe_df_melt = whe_df_local.melt(var_name='With Exercise Likert', value_name='Rating')
    
    # creates normalised table
    td_table = pd.crosstab(td_df_melt['Typical Day Likert'], td_df_melt['Rating'], normalize='index')*100
    
    # sorts for rows and cols
    row_order = ['Energy', 'Concentration', 'Productivity', 'Stress']  
    col_order = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
    
    # Sort the table based on the order
    td_table = td_table.reindex(col_order, axis=1)

    print('Table showing percentage of respondents for the typical day likert questions')
    display(td_table.loc[row_order].astype(int))
    
    # creates normalised table
    whe_table = pd.crosstab(whe_df_melt['With Exercise Likert'], whe_df_melt['Rating'], normalize='index')*100
    
    # Sort the table based on the order
    whe_table = whe_table.reindex(col_order, axis=1)

    print('\nTable showing percentage of respondents for the working hours exercise likert questions')
    display(whe_table.loc[row_order].astype(int))
```

```python
# I used the following for reference [33],[34]
# https://pypi.org/project/plot-likert/
# https://github.com/nmalkin/plot-likert/blob/release/docs/guide.ipynb
# cols for typical day likert scales 

def likert_summary():
    # sets the subplots
    fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,8))

    ## ax1 plot
    # cols for typical day likert scales
    td_cols = ['td_energy','td_concentration','td_productivity','td_stress']
    # creates new datafame just with the typical day likert columns
    td_df = likert_df[td_cols]

    # categories for likert
    categories = ['Very Low','Low','Moderate','High','Very High']

    # setting ax1 title
    ax1.set_title('Energy, Concentration, Productivity and Stress levels \non a typical day')

    # ax1 likert_plot
    plot_likert.plot_likert(td_df,plot_scale=categories,plot_percentage=True,
                            bar_labels=True, ax=ax1, legend=0)
    
    # changes labels to grey font
    for text in ax1.texts:
        text.set_color('darkslategrey')


    ## ax2 plot
    # cols for work hour exercise likert scales
    whe_cols = ['whe_energy','whe_concentration','whe_productivity','whe_stress']
    # creates new datafame just with the work hour exercise likert columns
    whe_df = likert_df[whe_cols]

    ax2.set_title(('Energy, Concentration, Productivity and Stress levels '
                   '\non days when respondent exercised during work hours'))

    # ax2 likert_plot
    plot_likert.plot_likert(whe_df,plot_scale=categories,plot_percentage=True,
                            bar_labels=True, ax=ax2)
    
    # changes labels to grey font
    for text in ax2.texts:
        text.set_color('darkslategrey')

    # suppresses axis labels on ax2
    ax2.set_yticks([])
    ax2.set_yticklabels([])

    # reduces space between plots
    plt.subplots_adjust(wspace=0.1)

    # changes names of y labels to be more interpretable
    ax1.set_yticks([3,2,1,0])
    ax1.set_yticklabels(['Energy','Concentration','Productivity','Stress'])

    plt.show()
```

```python
# getting paired ttest stats [35]
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html
energy_pairedt = stats.ttest_rel(likert_df['td_energy_num'], likert_df['whe_energy_num'])
concentration_pairedt = stats.ttest_rel(likert_df['td_concentration_num'], likert_df['whe_concentration_num'])
productivity_pairedt = stats.ttest_rel(likert_df['td_productivity_num'], likert_df['whe_productivity_num'])
stress_pairedt = stats.ttest_rel(likert_df['td_stress_num'], likert_df['whe_stress_num'])

# function for analysing. Dividing p-value by 2 as one tailed
def likert_ttest(ttest):
    if ttest[1]/2 < 0.05:
        print(f'The statistic is {ttest[0].round(4)} with a p-value of {ttest[1].round(4)/2}\n'
             'As it is below 0.05 this is a statistically significant result')
    else:
        print(f'The statistic is {ttest[0].round(4)} with a p-value of {ttest[1].round(4)/2}\n'
             'As it not below 0.05 this is not a statistically significant result')
```

```python
print('Paired t-test result for Energy Likert Scale\n')
likert_ttest(energy_pairedt)
print('\nPaired t-test result for Concentration Likert Scale\n')
likert_ttest(concentration_pairedt)
print('\nPaired t-test result for Productivity Likert Scale\n')
likert_ttest(productivity_pairedt)
print('\nPaired t-test result for Stress Likert Scale\n')
likert_ttest(stress_pairedt)
```

## Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
